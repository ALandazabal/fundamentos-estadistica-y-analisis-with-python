```python
import scipy.stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from scipy.stats import binom

p = 0.3
n = 8

x = list(range(0,9,1))
y = list([])

for i in x:
    y.append( binom.pmf(i, p = p, n = n) )
    
fig, ax = plt.subplots()
ax.bar(x,y)
ax.set_ylabel('Probabilidad discreta')

np.sum(y)
```

```python
media = np.average(x, weights = y)

varianza = np.average( (( x - media )**2), weights = y )

media, varianza

fig, ax = plt.subplots()
ax.bar(x,y)
ax.set_ylabel('Probabilidad discreta')
ax.axvline( x = media, c = 'r', linestyle = '--', label = 'Valor esperado')
ax.axvline( x = media + 3*np.sqrt(varianza), c = 'r', linestyle = '--', label = 'Desviaci√≥n estandar')
ax.legend()
```
